Ahora vamosa revisar las interaciones para ver si efectivamente tienen algun tipo de correlacion

```python
import statsmodels.formula.api as sm
#import chart_studio.plotly     as py
#pip install plotly
```

```python
print(df1.columns)
```

```python
model1 = 'profit~quantity  + ship_mode + order_priority + discount + segment '
lm1   = sm.ols(formula = model1, data = df1).fit()
print(lm1.summary())
```

Revisando los resultados vemos que la cantidad y descuento son variables que aparecen siginificativas pero el aporte al modelo es muy pequeÃ±o

```python
model1 = 'profit~discount + year'
lm1   = sm.ols(formula = model1, data = df1).fit()
print(lm1.summary())
```

La conclucion es que como tal este modelo no es el indicado para predecir puesto que las variables que tenemos tienen muy poco aporte y eso lo podemos ver en el R2 que tiene un score de
 0.11 y para poder considerar un score bueno debe estar arriba de los 0.70
